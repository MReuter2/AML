```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
np.random.seed(1)
```

```python
# multiply and add by random numbers to get some real values
# randn generates samples from the normal distribution (important - see below)
random_data = np.random.randn(50000)  * 20 + 20
```

```python
sns.boxplot(data=random_data)
plt.show()
```

## TASK

```python
# Import all datasets
import pandas as pd

spotify = pd.read_csv("./data/kaggle/Most Streamed Spotify Songs 2024.csv", thousands=',', encoding='ISO-8859-1')
height = pd.read_csv("./data/kaggle/SOCR-HeightWeight.csv")
birthrate = pd.read_csv("./data/kaggle/europe.csv")
```

```python
streams = spotify['Spotify Streams']
heights = height['Height(Inches)']
birthrates = birthrate['birth_rate']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(ax=axes[0], data=streams)
axes[0].set_title('Streams')

sns.boxplot(ax=axes[1], data=heights)
axes[1].set_title('Heights')

sns.boxplot(ax=axes[2], data=birthrates)
axes[2].set_title('Birth rates')

plt.show()
```

As can be seen, the number of streams on Spotify is likely not normally distributed, as the median lies clearly in the lower part of the box. Furthermore, several potential outliers are visible above the box. In contrast, the other boxplots show medians located near the center of the boxes, suggesting that these datasets (height and birth rate) are closer to a normal distribution. This is further supported by the relatively symmetrical distribution of potential outliers above and below the boxes in these cases.

## How can you test your data for being normally distributed?
The boxplots can only be used to visually test whether a data set is normally distributed. Further statistical tests must be carried out to clearly determine a normal distribution.
